4. Визуализация распределения переменных

```python
#Гистограмма для числовой переменной number_project (тип int64)
sns.histplot(df['number_project'], kde=True) #kde=True - кривая плотности распределения
plt.xlabel('number_project')
plt.ylabel('Частота')
plt.show()
```

```python
#Столбчатая диаграмма для категориальной переменной department:
sns.countplot(x='department', data=df)
plt.title('Распределение department')
plt.ylabel('Количество')
plt.show()
```

```python
#Боксплот для числовой переменной satisfaction_level:
sns.boxplot(x=df['satisfaction_level'])
plt.title('Boxplot satisfaction_level')
plt.xlabel('satisfaction_level')
plt.show()
```

Наблюдения: Медиана расположена между 0.6 и 0.8

Большая часть работников - недовольна, малая - наоборот
